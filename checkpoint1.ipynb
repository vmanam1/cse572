```python
import pandas as pd

df = pd.read_csv(r"V:\f25\cse572\project\dataset\customer_churn_dataset.csv")

print(df.shape)
print(df.head())
print(df.info())
print(df['Churn'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['Churn'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Stayed','Churned'], colors=['green','red'])
plt.title("Churn Distribution")
plt.ylabel("")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Contract Length', hue='Churn')
plt.title("Contract Length vs Churn")
plt.show()
```

```python
sns.countplot(data=df, x='Subscription Type', hue='Churn')
plt.title("Subscription Type vs Churn")
plt.show()
```

```python
sns.histplot(data=df, x='Tenure', hue='Churn', bins=30, kde=False)
plt.title("Tenure Distribution by Churn")
plt.show()
```

```python
sns.boxplot(data=df, x='Churn', y='Total Spend')
plt.title("Total Spend vs Churn")
plt.show()
```

```python
sns.countplot(data=df, x='Support Calls', hue='Churn')
plt.title("Support Calls vs Churn")
plt.xlabel("Number of Support Calls")
plt.ylabel("Count")
plt.show()
```
